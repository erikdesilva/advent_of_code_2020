```python
import os
from re import findall

cwd = os.getcwd()
input_txt = os.path.join(cwd, 'input.txt')

with open(input_txt, newline='\n') as input_file:
    lst_data = ''.join(input_file.read()).split('\n\n')
    lst_data = [i.replace('\n', ' ') for i in lst_data]
```

```python
my_str = 'ecl:grn hcl:#6b5442 cid:317 byr:2001 eyr:2023 iyr:2016 pid:407685013 hgt:177cm'
```

```python
passport_params = {
    'byr': [1920, 2002],
    'iyr': [2010, 2020],
    'eyr': [2020, 2030],
    'hgt': {'cm': [150, 193],
           'in': [59, 76]},
    'hcl': '#[0-9]{6}',
    'ecl': ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth'],
    'pid': '[0-9]{9}'
}
```

```python
def string_to_dict(string):
    tuples = findall('([a-z]{3}):([a-z0-9]+)', string)
    passport_dict = {}
    for item in tuples:
        passport_dict[item[0]] = item[1]

    return passport_dict
```

```python
valid_count = 0
```

```python
for row in lst_data:
    req_fields = passport_params.keys()
    fields_present = []
    pass_fields = findall('[a-z]{3}:', row)
    pass_fields = [i.rstrip(':') for i in pass_fields]
    for field in req_fields:
        if field in pass_fields:
            fields_present += [True]
        else:
            fields_present += [False]
    print(fields_present)
    if all(fields_present) == True:
        valid_count += 1
```
